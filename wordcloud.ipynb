```
data=pd.read_csv("/gdrive/My Drive/sentiment_dtm_train.csv",encoding="CP949")
del data['Id']
```

```
mask1 = (data.label == 1)
pos = data.loc[mask1,:].sum(axis=0)
del pos['label']
pos_dict = pos.to_dict()

mask2 = (data.label == 0)
neg = data.loc[mask2,:].sum(axis=0)
del neg['label']
neg_dict = neg.to_dict()
```

```
neg_dict
```

```
from wordcloud import WordCloud 

import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
%matplotlib inline

import matplotlib
from IPython.display import set_matplotlib_formats
matplotlib.rc('font',family = 'Malgun Gothic')

set_matplotlib_formats('retina')

matplotlib.rc('axes',unicode_minus = False)
```

```
wordcloud = WordCloud(font_path = 'drive/My Drive/Typo_DabangguB.ttf', 
                      background_color='white',colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(pos_dict)

plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```

```
wordcloud = WordCloud(font_path = 'drive/My Drive/Typo_DabangguB.ttf', 
                      background_color='white',colormap = "Accent_r", width=1500, height=1000).generate_from_frequencies(neg_dict)

plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```
